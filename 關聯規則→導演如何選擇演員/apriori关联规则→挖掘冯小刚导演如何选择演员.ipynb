```python
from efficient_apriori import apriori
import csv
```

```python
#文件名称
file='冯小刚.csv'
#读取文件
lists=csv.reader(open(file,'r',encoding='utf-8-sig'))
#数据加载
data=[]
for names in lists:
    name_new=[]
    for name in names:
        #去掉演员数据中的空格
        name_new.append(name.strip())
    data.append(name_new[1:])
```

```python
#挖掘频繁項集和关联规则

#支持度指的是某个商品组合出现的次数与总次数之间的比例。支持度越高，代表这个组合出现的概率越大
#置信度是一个条件概念，就是在 A 发生的情况下，B 发生的概率是多少
#提升度代表的是“商品 A 的出现，对商品 B 的出现概率提升了多少”

#min_support 参数为最小支持度,min_confidence 是最小置信度
itemsets, rules=apriori(data,min_support=0.1,min_confidence=1)
print(itemsets)
print(rules)
```

```python
#当支持度为5%
itemsets, rules=apriori(data,min_support=0.05,min_confidence=1)
print(itemsets)
print(rules)
```
